```python
# The SIR model, one time step - PA First
def sir(y, PAbeta, gamma, N):
    PAS, PAI, PAR = y
    PASn = (-PAbeta * PAS * PAI) + PAS
    PAIn = (PAbeta * PAS * PAI - gamma * PAI) + PAI
    PARn = gamma * PAI + PAR
    if PASn < 0:
        PASn = 0
    if PAIn < 0:
        PAIn = 0
    if PARn < 0:
        PARn = 0

    scale = N / (PASn + PAIn + PARn)
    return PASn * scale, PAIn * scale, PARn * scale
```

```python
def sim_sir(PAS, PAI, PAR, PAbeta, gamma, n_days, beta_decay=None):
    N = PAS + PAI + PAR
    PAs, PAi, PAr = [PAS], [PAI], [PAR]
    for day in range(n_days):
        y = PAS, PAI, PAR
        PAS, PAI, PAR = sir(y, PAbeta, gamma, N)
        if beta_decay:
            PAbeta = PAbeta * (1 - beta_decay)
        PAs.append(PAS)
        PAi.append(PAI)
        PAr.append(PAR)

    PAs, PAi, PAr = np.array(PAs), np.array(PAi), np.array(PAr)
    return PAs, PAi, PAr
```
